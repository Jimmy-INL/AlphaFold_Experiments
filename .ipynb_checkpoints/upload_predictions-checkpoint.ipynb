```python
folder = '/home/xander/data/'
network = 'no-potts'
```

```python
import os, shutil
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl  
import pandas as pd

def getListOfFiles(dirName):
    if dirName[-1] != '/': dirName += '/'
    listOfFile = os.listdir(dirName)
    allFiles = list()
    for entry in listOfFile:
        fullPath = os.path.join(dirName, entry)
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)
                
    return sorted(allFiles)

def load(fileName):
    fileObject2 = open(fileName, 'rb') 
    modelInput = pkl.load(fileObject2) 
    fileObject2.close() 
    return modelInput

def plot_distogram(prediction, domain_id, network, stride, savepath):
    plt.figure()
    plt.imshow(-np.argmax(prediction[1:,:,:],axis=0))
    plt.suptitle('ProSPr '+domain_id+' Prediction', fontsize=14)
    plt.title('Network '+network+', Stride '+stride, fontsize=10)
    plt.xlabel('Residue i')
    plt.ylabel('Residue j')
    plt.savefig(savepath)
    plt.plot()
```

```python
prediction_files = [f for f in getListOfFiles(folder) if 'predictions.pkl' in f]
print('Found %d distograms in %s:\n' %(len(prediction_files), folder))
for f in prediction_files:
    print(f)
```

```python
for prediction_file in prediction_files:
    prediction = load(prediction_file)
    domain_id, stride = prediction_file.split('_predictions.pkl')[0].split('/')[-1].split('__')
    
    savepath = os.path.join(os.path.dirname(prediction_file), '%s_distogram_plot.jpg' %domain_id)
    plot_distogram(prediction, domain_id, network, stride, savepath)
```
