```python
fasta_sequence_file = '/home/xander/data/zoo/zoo.fasta'
```

```python
def parse_fasta_file(filepath):
    sequences = {}
    with open(filepath) as fp:
        line = fp.readline()
        cnt = 1
        while line:
            line_text = line.strip()
            if len(line_text) > 0:
                if line_text[0] == '>':
                    current_sequence = line_text[1:]
                    sequences[current_sequence] = ''
                else:
                    sequences[current_sequence] = line_text
            line = fp.readline()
            cnt += 1
    return sequences
        
sequence_dict = parse_fasta_file(fasta_sequence_file)
```

```python
import os, shutil

def create_fasta_file(sequence_name, sequence):
    fasta_dir      = '/home/xander/data/%s' %sequence_name
    fasta_filepath = os.path.join(fasta_dir, '%s.fasta' % sequence_name)
    
    shutil.rmtree(fasta_dir)
    os.makedirs(fasta_dir, exist_ok = True)
    with open(fasta_filepath, 'a') as fastafile:
        fastafile.write('>%s\n' %sequence_name)
        fastafile.write('%s' %sequence)


n = 0
for sequence_name, sequence in sequence_dict.items():
    print('%16s: %s' %(sequence_name, sequence))
    create_fasta_file(sequence_name, sequence)
    n+=1
    
print("\nFound a total of %d sequences!" %n)
```

## Get predictions using alphafold:

```python
import subprocess

def get_command(sequence_name, task, no_potts=True):
    cmd = ['sudo', 'docker', 'run', '-it', '-v', '/home/xander/data:/data', 'xander_prospr',
           '%s' %task, '%s' %sequence_name]
    
    if no_potts:
        cmd = cmd + ['--network=no-potts']
        
    return cmd
```

```python
script_path = 'run_alphafold.sh'
try:
    os.remove(script_path)
except:
    print("No script with name %s was found... Creating new one!" %script_path)

with open(script_path, 'a') as shell_script:
    
    for sequence_name, sequence in sequence_dict.items():
        cmd = get_command(sequence_name, 'build', no_potts=True)
        #print("Running: %s" % ' '.join(cmd))
        #p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        #out, err = p.communicate()
        shell_script.write('%s\n' %(' '.join(cmd)))

        cmd = get_command(sequence_name, 'run', no_potts=True)
        #print("Running: %s" % ' '.join(cmd))
        #p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        #out, err = p.communicate()
        
        shell_script.write('%s\n' %(' '.join(cmd)))
```
